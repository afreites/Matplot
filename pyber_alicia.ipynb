```python
# Instructor Demo
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# File to Load (Remember to change these)
city_data_path= "city_data.csv"
ride_data_path= "ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_path)
ride_data= pd.read_csv(ride_data_path)
```

```python
# Combine the data into a single dataset
pyber_df = pd.merge(ride_data, city_data, how='left', on='city')
pyber_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
rural_df=pyber_df.loc[pyber_df['type'] == 'Rural']
rural_df.head()
```

```python
rural_df.describe()
```

```python
city_count_rural=rural_df.groupby('city')['ride_id'].count() 
```

```python
avg_fare_rural=rural_df.groupby('city')['fare'].mean() 
```

```python
city_driver_rural=rural_df.groupby('city')['driver_count'].mean() 
```

```python
suburban_df=pyber_df.loc[pyber_df['type'] == 'Suburban']
suburban_df.head()
```

```python
suburban_df.describe()
```

```python
city_count_suburban=suburban_df.groupby('city')['ride_id'].count() 
```

```python
avg_fare_suburban=suburban_df.groupby('city')['fare'].mean() 
```

```python
city_driver_suburban=suburban_df.groupby('city')['driver_count'].mean() 
```

```python
urban_df=pyber_df.loc[pyber_df['type'] == 'Urban']
urban_df.head()
```

```python
urban_df.describe()
```

```python
city_count_urban=urban_df.groupby('city')['ride_id'].count() 
```

```python
avg_fare_urban=urban_df.groupby('city')['fare'].mean() 
```

```python
city_driver_urban=urban_df.groupby('city')['driver_count'].mean() 
```

```python
fig, ax = plt.subplots()
plt.scatter(city_count_rural, avg_fare_rural,marker="o",facecolors="gold",alpha=0.75, edgecolors="gold", s=city_driver_rural*5,label="Rural")
plt.scatter(city_count_suburban, avg_fare_suburban, marker="o", facecolors="lightblue",alpha=0.75,edgecolors="lightblue", s=city_driver_suburban*8,label="Suburban")
plt.scatter(city_count_urban, avg_fare_urban, marker="o", facecolors="lightcoral", alpha=0.75,edgecolors="lightcoral", s=city_driver_urban*12,label="Urban")
plt.title("Pyber Ride Sharing Data 2016 ")
plt.xlabel("Number of Rides ( per City)")
plt.ylabel("Average Fare ($)")
plt.ylim(18,45)
plt.xlim(3,41)
ax.set_frame_on(True)
ax.grid(linestyle='-', linewidth='0.5', color='gray')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left',frameon=False, borderaxespad=0.)
textstr ='Note: \n Circles represent amount of drivers'
# place a text box in upper left in axes coords
ax.text(1.03,.72, textstr, transform=ax.transAxes, fontsize=8,
        verticalalignment='top')
plt.tight_layout()
plt.savefig('/Users/afreites/anaconda3/envs/pythondata/BubbleGraph.pdf')
plt.savefig('BubbleGraph.pdf', format='pdf')
plt.show()
```
